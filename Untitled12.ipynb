```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error

# Creating a synthetic "messy" dataset for demonstration
data = {
    'car_id': [1, 2, 3, 4, 5, 2],
    'brand': [' Toyota ', 'honda', 'TOYOTA', 'Ford', 'nissan', 'honda'],
    'mileage': ['15000 km', '25000km', np.nan, '5000 miles', '12000', '25000km'],
    'selling_price': [15000, 12000, 18000, np.nan, 9000, 12000]
}
df = pd.DataFrame(data)

# --- Task 1: Explore and Identify Issues ---
print("Dataset Info:")
df.info()
print("\nDescriptive Stats:")
print(df.describe())
print("\nShape:", df.shape)

# Identified Issues:
# 1. Missing Values: 'selling_price' (target) and 'mileage' have NaNs.
# 2. Data Types: 'mileage' is an object (string) instead of numeric.
# 3. Inconsistent Strings: 'brand' has mixed casing and leading/trailing spaces.
# 4. Duplicates: Row index 1 and 5 are identical.

# --- Task 2: Clean the Data ---

# 1. Drop rows where the target (selling_price) is missing
df = df.dropna(subset=['selling_price'])

# 2. Extract numeric values from mileage and convert to float
# This removes 'km', 'miles', and spaces
df['mileage'] = df['mileage'].str.extract('(\d+)').astype(float)

# 3. Impute missing mileage with the median
df['mileage'] = df['mileage'].fillna(df['mileage'].median())

# 4. Strip and lowercase the brand column
df['brand'] = df['brand'].str.strip().str.lower()

# 5. Remove duplicates
df = df.drop_duplicates()

print("\n--- Cleaned Dataframe ---")
print(df)

# --- Task 3: Compute Baseline MAE ---

# The baseline prediction is simply the mean of the actual selling prices
baseline_prediction = df['selling_price'].mean()

# Create a list of predictions (the same mean value for every row)
y_true = df['selling_price']
y_pred = [baseline_prediction] * len(df)

mae = mean_absolute_error(y_true, y_pred)

print(f"\nBaseline Mean Price: ${baseline_prediction:.2f}")
print(f"Baseline Mean Absolute Error: {mae:.2f}")
```
